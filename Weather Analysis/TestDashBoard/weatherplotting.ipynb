```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("CleanedWeatherData.csv")
```

```python
df
```

```python
sd=df.sort_values(['temperature_celsius'],ascending=0).head(10)
```

```python
sd
```

```python
sd.groupby(['temperature_celsius','country']).count()
```

```python
grouped_df = df.groupby('country', as_index=False).agg({'temperature_celsius': 'mean'})
```

```python
sorted_df=grouped_df.sort_values(['temperature_celsius'],ascending=0)
```

```python
sorted_df
```

```python
colors = plt.cm.tab10(range(len(sorted_df)))
```

```python

# Plotting the data
plt.figure(figsize=(10, 5))
plt.bar(sorted_df['country'].head(10), sorted_df['temperature_celsius'].head(10), color=colors)
plt.ylabel('Temperature (Celsius)')
plt.xlabel('Location')
plt.title('Temperature in Different Locations')
plt.yticks(rotation=90)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Plotting a pie chart
plt.figure(figsize=(10, 10))
plt.pie(sorted_df['temperature_celsius'].head(10), labels=sorted_df['country'].head(10), autopct='%1.2f%%', startangle=130,colors=colors)
plt.title('Distribution of Average Temperature Across Top 10 Countries(Highest temperature)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
colors = plt.cm.tab10(range(len(sorted_df)))

# Plotting a scatter plot with custom colors
plt.figure(figsize=(10, 6))
plt.scatter(sorted_df['country'].head(10), sorted_df['temperature_celsius'].head(10), s=100, alpha=0.8, edgecolors='k')
plt.xlabel('Country')
plt.ylabel('Average Temperature')
plt.title('Scatter Plot:Distribution of Average Temperature Across Top 10 Countries(Highest temperature)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
groupeds_df = df.groupby('country', as_index=False).agg({'temperature_celsius': 'mean'})
sorteds_df=grouped_df.sort_values(['temperature_celsius'])
plt.figure(figsize=(20, 7))
plt.bar(sorteds_df['country'].head(10), sorteds_df['temperature_celsius'].head(10), color=colors)
plt.ylabel('Temperature (Celsius)')
plt.xlabel('Location')
plt.title('Temperature in Different Locations')
plt.yticks(rotation=90)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Plotting a pie chart
plt.figure(figsize=(10, 10))
plt.pie(sorteds_df['temperature_celsius'].head(10), labels=sorteds_df['country'].head(10), autopct='%1.1f%%', startangle=140,colors=colors)
plt.title('Distribution of Average Temperature Across Top 10 Countries(Lowest temperature)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
