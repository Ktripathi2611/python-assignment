```python



plt.bar(df['name'].value_counts().index, df['name'].value_counts().values)
plt.xlabel('Pizza type')
plt.ylabel('Count of pizza')
plt.show()
```

```python


plt.figure(figsize=[40,20])
name_counts = df['name'].value_counts()
plt.bar(name_counts.index, name_counts.values)
plt.xlabel('Name')
plt.ylabel('Count')
plt.title('Name Frequencies')
plt.show()
```

```python
plt.figure(figsize=[20,10])
plt.barh(df['name'].value_counts().index, df['name'].value_counts().values)
plt.xlabel('Pizza type')
plt.ylabel('Count of pizza')
plt.show()
```

```python


plt.figure(figsize=(8, 8))
plt.pie(df['name'].value_counts(), labels=df['name'].value_counts().index, autopct='%1.1f%%', startangle=90)
plt.title('Pizza Types Distribution')
plt.axis('equal')  
plt.show()
```
